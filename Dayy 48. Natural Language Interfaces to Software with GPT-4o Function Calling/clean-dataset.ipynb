```python
import pandas as pd
```

```python
sales = pd.read_csv("data-raw/Bakery Sales.csv", nrows=2420, parse_dates=["datetime"], usecols=["datetime", "croissant", "almond croissant", "tiramisu croissant", "pain au chocolat"]) \
    .query('datetime >= "2019-08-01" & datetime < "2020-01-01"') \
    .groupby(pd.Grouper(key="datetime", freq="1D")) \
    .sum() \
    .reset_index() \
    .melt(id_vars="datetime", var_name="product", value_name="n_sold")
sales
```

```python
sales.to_csv("bakery_sales.csv", index=False)
```

```python
def get_total_sales(product, month_name):
    return sales \
        .query('product == @product & datetime.dt.month_name() == @month_name') \
        ["n_sold"] \
        .sum()
get_total_sales("croissant", "August")
```
